```python
fig = plt.figure(figsize=(8, 6))  # Specify figure size in inches
# Add a single subplot to the figure
ax = fig.add_subplot(111)  # (1 row, 1 column, 1st subplot)
# Display the image
ax.imshow(img_rgb)
ax.axis('off')  # Turn off axis labels and ticks
ax.set_title("Image Display")  # Set a title for the subplot
# Show the figure
plt.show()
```

```python
cv2.line(img_rgb, (50, 50), (200, 50), (255, 0, 0), 5)
cv2.rectangle(img_rgb, (60, 60), (160, 160), (0, 255, 0), 3)
cv2.circle(img_rgb, (120, 120), 40, (0, 0, 255), -1)
plt.imshow(img_rgb)
plt.show()
```

```python
img1 = cv2.imread('dog.jpg')
```

```python
img_rgb1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb1)
plt.show()
```

```python
# Draw a circle
cv2.circle(img_rgb1, (15, 15),20, (255, 0, 0), -1)  # Center, Radius, Color, Thickness
plt.imshow(img_rgb1)
plt.show()

```

```python
# Draw a line
line_image = img1.copy()
cv2.line(line_image, (50, 50), (200, 200), (0, 255, 0), 2)  # Start, End, Color, Thickness
plt.imshow(line_image)
plt.show()
```

```python
# Draw a rectangle
rect_image = img1.copy()
cv2.rectangle(rect_image, (20, 20), (150,100), (0, 0, 255), 2)  # Top-left, Bottom-right, Color, Thickness
plt.imshow(rect_image)
plt.show()

```

```python
shapes_image = img1.copy()
cv2.circle(shapes_image, (100, 100), 50, (255, 0, 0), 2)
cv2.line(shapes_image, (50, 50), (200, 200), (0, 255, 0), 2)
cv2.rectangle(shapes_image, (15, 15), (150, 150), (0, 0, 255), 2)
plt.imshow(shapes_image)
plt.show()
```

```python
plt.imshow(img1)
plt.imshow('Top-Left', img1[:10, :10])  # Crop top-left
plt.imshow('Top-Right', img1[:100, -100:])  # Crop top-right
plt.show()
```
